
# **Project Goal**
This project focuses on studying and forecasting the productivity of garment factory employees using data mining methods. The main objective is to determine whether employees can reach their productivity targets and to group them according to similar performance patterns. By applying both classification and clustering techniques, the project aims to analyze employee data, estimate the probability of achieving productivity goals, and uncover meaningful patterns among workers with shared characteristics.


Classification:

In the classification phase, a predictive model will be built to decide whether an employee is likely to achieve the expected productivity level. The model will rely on factors such as working hours, experience, overtime, incentives, and other relevant attributes. The purpose is to identify which employees are most likely to meet their productivity goals based on their profiles.

Clustering:
In the clustering phase, employees will be organized into groups with similar productivity behaviors. These groups will help in predicting outcomes for new employees by comparing them to the most similar cluster, which supports better understanding and management of employee performance.
